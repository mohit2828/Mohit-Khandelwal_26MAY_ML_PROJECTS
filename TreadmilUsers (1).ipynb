```python
import matplotlib.pyplot as plt
```

```python
for i in ['Product','MaritalStatus','Gender']:
    plt.figure()
    df['Product'].value_counts().plot.pie(autopct='%.1f', figsize=(8,8))
    plt.title('Pie Chart of {}s'.format(i) )
```

```python
for i in ['Product','MaritalStatus','Gender']:
    df['Product'].value_counts().plot.pie(autopct='%.1f', figsize=(8,8))
    plt.title('Pie Chart of {}s'.format(i) )
    plt.show()
```
